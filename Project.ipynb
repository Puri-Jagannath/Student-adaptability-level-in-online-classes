# Label Encoding

```python
data = df

all_c = data.columns

for column in all_c:
    print(data[column].value_counts())
    print(60*'-')
```

```python

d1 = {'Yes' : 0, 'No' : 1}
data['IT Student'] = data['IT Student'].map(d1)

d2 = {'Low' : 1, 'High' : 0 }
data['Load-shedding'] = data['Load-shedding'].map(d2)

d3 = {'Boy' : 0, 'Girl' : 1}
data['Gender'] = data['Gender'].map(d3)

d4 = {'Wifi' : 1, 'Mobile Data' : 0}
data['Internet Type'] = data['Internet Type'].map(d4)

d5 = {'Mobile' : 0, 'Computer' : 1, 'Tab' : 2}
data['Device'] = data['Device'].map(d5)

d6 = {'Non Government' : 0, 'Government' : 1}
data['Institution Type'] = data["Institution Type"].map(d6)

d7 = {'School' : 0, 'University' : 1, 'College' : 2}
data['Education Level'] = data['Education Level'].map(d7)

d8 = {'Poor' : 1, 'Mid' : 0, 'Rich' : 2}
data['Financial Condition'] = data['Financial Condition'].map(d8)

d9 = {'4G' : 0, '3G' : 1, '2G' : 2}
data['Network Type'] = data['Network Type'].map(d9)

d10 = {'21-25' : 0, '11-15' : 1, '16-20' : 2, '1-5' : 3, '26-30' : 4, '6-10' : 5}
data['Age'] = data['Age'].map(d10)

d11 = {'0' : 2, '1-3' : 0, '3-6' : 1}
data['Class Duration'] = data['Class Duration'].map(d11)

d12 = {'Low' : 0, 'Moderate' : 1, 'High' : 2}
data['Adaptivity Level'] = data['Adaptivity Level'].map(d12)

d14 = {'Yes' : 1, 'No' : 0}
data['Self Lms'] = data['Self Lms'].map(d1)
```

# Building a Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score
from sklearn.linear_model import LogisticRegression
```

```python
X = data.drop(columns="Adaptivity Level")           
y = data["Adaptivity Level"] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
LR = LogisticRegression()
LR.fit(X_train, y_train)
ypred = LR.predict(X_test)
print(accuracy_score(y_test, ypred)*100)
```
